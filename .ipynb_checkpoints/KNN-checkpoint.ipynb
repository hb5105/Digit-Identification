```python
#step1 is to find k nearest neighbours 
#based oon eucledean distance
```

```python
#distance formula
def distance(x1,x2):
    return np.sqrt(sum(x1-x2)**2)
```

```python
#KNN ALGO
def knn(x,y,querypoint,k=5):
    #pick the k nearest neighbours
    
    vals=[]
    for i in range(x.shape[0]):
        #compute the distances 
        d=distance(querypoint,x[i])
        vals.append((d,y[i]))
        
    #print(vals)
    # sort the vals
    vals=sorted(vals)
    vals=vals[:k]
    
    vals=np.array(vals)
    new_v = np.unique(vals[:,1],return_counts=True)
    print(new_v)
    
    index = new_v[1].argmax()
    pred = new_v[0][index]
    
    return pred
    
```

```python
knn(x,y,[4,4]) #gives prediction using knn
```
